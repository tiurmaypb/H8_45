```python
alice_wordcloud= WordCloud(background_color="white",
                          max_words=2000,
                          stopwords=STOPWORDS)
alice_wordcloud.generate(alice)
```

```python
plt.imshow(alice_wordcloud)
plt.show
```

```python
plt.figure(figsize=(10,5))
plt.imshow(alice_wordcloud,interpolation ="bilinear")
plt.axis("off")
plt.show
```
